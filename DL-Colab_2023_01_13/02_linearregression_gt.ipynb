# 데이터 준비

```python
x_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24])
```

### 경사하강법

```python
# 초기화

W = tf.Variable(0.0)
# W.numpy() # tensorflow를 넘파이로 바꾸는 함수.
b = tf.Variable(0.0) # GPU에서 빠르게 동작하려면 tensorflow가 원하는 객체형태로 만들어주는 것이 좋다.

y_hat = W * x_train + b
```

```python
lr = 0.01

from tensorflow.python.eager.backprop import GradientTape
# MSE
cost = tf.reduce_mean(tf.square(y_train - y_hat)) # square은 제곱, square는 평균, reduce는 차원감소, 2차원일경우 1차원으로 바꿈.

# 자동미분 기능
with tf.GradientTape() as tape:
  y_hat = W * x_train + b
  cost = tf.reduce_mean(tf.square(y_train - y_hat))

W_grad, b_grad = tape.gradient(cost, [W, b])

# W = W - lr * W_grad
W.assign_sub(lr * W_grad)
b.assign_sub(lr * b_grad)
```

```python
W.numpy(), b.numpy()
```

### 전체코드

```python

W = tf.Variable(0.0)
b = tf.Variable(0.0)
epochs = 5000
lr = 0.01

for i in range(epochs):
  with tf.GradientTape() as tape:
    y_hat = W * x_train + b
    cost = tf.reduce_mean(tf.square(y_train - y_hat))

    W_grad, b_grad = tape.gradient(cost, [W, b])

    # W = W - lr * W_grad
    W.assign_sub(lr * W_grad)
    b.assign_sub(lr * b_grad)

    if i % 200 == 0:
      print(f'Epochs:{i}, cost:{cost.numpy():10f}, W:{W.numpy():10f}, b:{b.numpy():10f}')
```
