# 선형회귀 (Dense Layer)
---

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# 데이터 준비

```python
x_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 12., 15., 18., 21., 24])
```

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(units=1, input_shape=(1,)), # 완전 연결층 (Fully Connected Layer) # tensorflow에서는 dense layer이라 하고, pytorch에서는 linear layer이라고 함.
])

# layer안에 unit이 있는데, 노드같이 생긴 것을 unit이라고 함. layer은 그 라인의 unit들을 이루는 층.
# 앞의 unit에서 뒤의 unit으로 데이터를 전송하는데, 데이터의 크기가 일치해야지 그 모델을
# 뒤의 unit에서 사용할 수 있다.
```

```python
model.summary() 
# Output Shape에서 (None, 1)인것은, None은 들어오는 데이터의 수를 알수없기 때문에 None이고, 1은 출력의 개수이다. unit의 개수가 하나라고 봐도된다.
# Param는 parameter 개수이다.
# Non-trainable params : 0, 다른 사람이 만들어둔, 학습시킨 모델을 가져오면 개수가 나타나지가 않음.
# Trainable params : 2, 우리가 학습시켜야 하는 데이터는 두개라는 뜻.
```

```python
# 컴파일

adam = tf.keras.optimizers.Adam(learning_rate = 0.01) # 이렇게 직접 optimizer을 설정하면 learning_rate를 설정할 수 있다.
sgd = tf.keras.optimizers.experimental.SGD()

model.compile( # 학습시키는 방법(컴파일)을 사용자가 지정하는 함수
    optimizer = sgd, # 경사하강법, gd같은 경사하강법의 종류를 정해주는 것, adam을 제일 많이 사용한다.
    # optimizer = 'adam'
    loss = 'mse', # mse, mean squared error
    metrics = ['mse'] # 어떤 학습지표를 보면서 진행할지
)
```

```python
# 학습

model.fit(x_train, y_train, 
          epochs=100,
          batch_size = 2,
          verbose = 1) # epochs는 출력횟수, 전에 코딩할때 epochs는 반복횟수였다.
# verbose는 0~3까지 입력이 가능한데, 큰 수를 입력할 수록 나타나는 데이터의 양이 많아진다.
```

```python
# 학습된 파라미터
model.get_weights()
```

```python
# 평가

model.evaluate(x_train, y_train) # 학습이 얼마나 진행되고, 학습이 어디에서 완료되었는지 확인이 가능하다.
```

```python
x_test = [10.]

y_pred = model.predict(x_test)
y_pred
```
